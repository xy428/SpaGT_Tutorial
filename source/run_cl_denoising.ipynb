```python
import os
os.environ['R_HOME']="/share/home/baixiaosheng/software/anaconda/envs/SpaGT/lib/R"
import scanpy as sc
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import SpaGT
SpaGT.__version__
```

```python
adata = sc.read_h5ad("./cl.h5ad")
```

```python
adata, conns = SpaGT.spatial_reconstruction(adata,alpha=1.0)
```

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
plot_gene = 'YBX1'
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
sc.pl.spatial(adata, img_key=None, color=plot_gene, show=False, ax=axs[0], title='RAW_'+plot_gene, vmax=2.5,size=1.5)

sc.pl.spatial(adata, img_key=None, color=plot_gene, show=False, ax=axs[1], title='Denoised_'+plot_gene, layer='ReX',vmax=2.5,size=1.5)
```
